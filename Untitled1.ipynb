```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list_losstr = np.loadtxt('D:\\Data\\Results\\wisig_siamkar\\wisig_siamkar\\list_losstr1.csv',delimiter=',')
list_losste = np.loadtxt('D:\\Data\\Results\\wisig_siamkar\\wisig_siamkar\\list_losste1.csv',delimiter=',')
```

```python
for splits in range(2):    
    #print("Iter:" + str(i))
    fig = plt.figure()
    ltr = [list_losstr[j][1] for j in range(len(list_losstr)) if list_losstr[j][0] == splits]
    lte = [list_losste[j][1] for j in range(len(list_losste)) if list_losste[j][0] == splits]
    plt.plot(list(np.arange(0,len(ltr),1)),ltr)
    plt.plot(list(np.arange(0,100*len(lte),100)),lte)
    plt.ylim([0,0.40])
    plt.title('loss plot')
    plt.show()
    plt.close(fig)
```

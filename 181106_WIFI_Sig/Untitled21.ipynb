```python
import os
import tensorflow as tf
import numpy as np
import pandas as pd
import pickle
import gzip
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# data path
path_csi =  'J:\\Data\\Wi-Fi_processed\\'
path_csi_hc = 'J:\\Data\\Wi-Fi_HC\\180_100\\'

# data info
df_info = pd.read_csv('data_subc_sig.csv')
# parameters
no_classes = np.max(df_info['id_person'])+1
epochs = 1
batch_size = 10
csi_time = int(np.max(df_info['len']))
csi_subc = 30
input_shape = (csi_time, csi_subc, 6)
```

```python
df_add = pd.DataFrame(columns=['id','median','max'])
for file in set(df_info.id.values):
    print(file)
    ser_add = pd.Series()
    # read sample data
    # load and uncompress.
    with gzip.open(path_csi+file+'.pickle.gz','rb') as f:
        data1 = pickle.load(f)
    data1_abs = np.abs(data1)
    ser_add['id']=file
    ser_add['median'] = np.median(data1_abs)
    ser_add['max'] = np.max(data1_abs)
    df_add = df_add.append(ser_add,ignore_index=True)
```

```python
df_add.to_csv('df_add.csv')
```

```python
df_info_new = pd.merge(df_info,df_add,how='left',on='id')
```

```python
df_info_new.to_csv('data_subc_sig_v1.csv',index=False)
```

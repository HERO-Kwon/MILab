```python
import numpy as np

def calcSma(data, smaPeriod):
    sma = []
    count = 0
    for i in range(data.size):
        if data[i] is None:
            sma.append(None)
        else:
            count += 1
            if count < smaPeriod:
                sma.append(None)
            else:
                sma.append(np.mean(data[i-smaPeriod+1:i+1]))

    return np.array(sma)
```

```python
plt.plot(calcSma(lstm1.Value.values,10),'r')
plt.plot(calcSma(lstm3[0],10),'b')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.ylim((0,5))
plt.legend(['training_loss','test_loss'])
```
